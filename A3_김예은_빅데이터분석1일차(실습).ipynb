### 데이터 결합 실습

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/home/piai/Downloads/FITNESS_1.csv', engine = 'python', encoding ="EUC-KR" )
df2 = pd.read_csv('/home/piai/Downloads/FITNESS_2.csv', engine = 'python', encoding ="EUC-KR")
```

```python
df1.head()
```

```python
df2.head()
```

```python
#NAME을 기준으로 두 dataframe merge하기
result = pd.merge(df1,df2,on='NAME')
result.head()
```

```python
#원본 FITNESS.CSV와 비교하여 동일하게 만들어주는 과정
# 필요없는 column들 삭제 및 column명 변경
del result['GENDER_y']
del result['AGE_y']
result.rename(columns = {result.columns[1]: 'GENDER',
                           result.columns[2]: 'AGE'}, inplace = True)

result.head()
```

```python
#GENDER column의 값 변경하기(방법1)
result.loc[result['GENDER'] == '남성', 'GENDER'] = 'M'
result.loc[result['GENDER'] == '여성', 'GENDER'] = 'F'
result.head()
```

```python
#GENDER column의 값 변경하기(방법2)
result['GENDER'] = result['GENDER'].str.replace('남성','M')
result['GENDER'] = result['GENDER'].str.replace('여성','F')
result.head()
```

```python
#AGEGROUP column의 값 변경하기
result['AGEGROUP'] = result['AGEGROUP'].str.replace('대','')
result.head()
```

### 데이터 변환

```python
df = pd.read_csv('/home/piai/Downloads/FITNESS.csv', engine = 'python', encoding ="utf-8" )
df.head()
```

```python
#GENDER column 값 변경
df['GENDER'] = df['GENDER'].str.replace('M','남성')
df['GENDER'] = df['GENDER'].str.replace('F','여성')
df.head()
```

```python
#GENDER , AGEGROUP, WEIGHT column 추출 후 WEIGHT의 평균 값 계산
df1 = df.groupby(['GENDER', 'AGEGROUP'], as_index = False)['WEIGHT'].agg('mean')
df1
```

```python
#결측치 제거
df1.drop([0], inplace = True)
df1
```

```python
#피벗사용해서 데이터 변환하기
df1.pivot(index = 'GENDER', columns = 'AGEGROUP', values ='WEIGHT')
```

```python
#피벗사용해서 데이터 변환하기
df1.pivot(index = 'AGEGROUP', columns = 'GENDER', values ='WEIGHT')
```
