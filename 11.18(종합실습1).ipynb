## 추진 배경 및 목적

1970년대 BOSTON 주택 가격과 그 가격에 영향을 주는 요인들에 대한 Data Set을 활용해 변수간 상관관계를 확인한다. 주택가격에 가장 큰 영향을 끼치는 데이터 수집/ 정제/ 변환, 분석계획 및 데이터 탐색적 분석(그래프로 시각화), 문제 분석 및 결과 해석(선형회귀, 피어슨), 모델 평가와 같은 지식들을 종합적으로 활용하고 실제적으로 체득하고자 한다.

데이터 분석을 통해 가설을 검정한 후, 주택 가격에 가장 큰 영향을 미치는 인자들의 정보를 추출한다. 분석한 정보를 토대로 집값안정화 정책 및 사업아이템 선정에 대한 인사이트를 도출해보고자 한다.

## 가설설정

목표변수는 MEDV(주택가격)이고, 여기에 영향을 미칠 것 같은 변수 5가지를 중심으로 가설을 설정해보았다.

가설1: 가구당 평균 방 수가 많으면 주택가격이 높을 것이다. 방 수가 많다는 것은 규모가 크다는 것을 의미하므로 주택가격이 더 높을 것이라고 생각했다. 만약 당시 사회적 트렌드가 소형집을 선호하는 것이었다면, 결과치가 다르게 나올 수도 있을 것이라고 생각한다. 

가설2: 중심지(직업센터) 접근 거리가 가까울수록 주택가격이 높을 것이다.
대부분의 사람들은 근무지와 주거지가 가까운 것을 선호하므로 양 쪽의 접근 거리가 가까울수록 주택가격이 높을 것이라고 예상했다. 하지만, 중심지의 소란스러움을 피해 교외로 나가는 경우도 있으므로 가설을 증명함으로써 알아보고 싶다. 

가설3: 학생/교사 비율이 낮을수록 주택가격이 높을 것이다.
교육의 질의 경우 학생 1명 당 교사 수가 많을 수록 높아진다고 판단했다. 따라서 학생/교사 비율이 낮을수록(=학생 1명당 교사 수가 많을수록) 주택에 대한 수요가 증가할 것이라고 생각했다. 하지만, 한편으로는 절대적인 학생 수가 많아 학생/교사 비율이 높은 경우도 고려해볼 수 있다. 이 경우에는 인구수가 많고, 지역에 대한 수요가 많아져 주택가격이 높아지는 것으로도 생각해볼 수 있다. 따라서 이 부분에 대한 상관관계를 파악해보고자 가설을 설정했다. 

가설4: 흑인 인구 비율과 저소득층 비율이 높은 지역은 범죄율이 높을 것이다. 1970년대 보스턴에는 저소득층 흑인들이 모여사는 슬럼가가 있었고, 범죄가 빈번하게 일어났었다고 한다. 범죄율이 높은 지역에 대한 주택 수요는 감소하므로, 그에 따라 집 값이 낮을 것이라고 생각했다.
