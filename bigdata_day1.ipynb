### 데이터 가져오기

```python
ds_cabbage = pd.read_csv('/home/piai/test/Daily/EXH_QC1.csv', engine = 'python')
ds_cabbage.head()
```

```python
#제품 결함과 결함이 발생한 기간의 데이터 subset
df = ds_cabbage[['Flaws','Period']]
df.head()
```

```python
#Flaws 변수 value count 순서대로 정렬 : sort_index()
count = ds_cabbage['Flaws'].value_counts().sort_index()
count
```

```python
#결함 percent 로 표현하기
percent = count / sum(count) * 100
percent
```

```python
# 누적 count , CumCnt (Cumulative counts) 계산 : cumsum
cumcnt = np.cumsum(count)
cumcnt
```

```python
# 누적 percent-> cumsum
cumpct = np.cumsum(percent)
cumpct
```

```python
#dataframe으로 취합
count_data = pd.DataFrame({'Count':count , 'Percent': percent, 
                          'Cumcnt':cumcnt,'Cumpct':cumpct})

#column name 지정
count_data.columns.name = 'Flaws'
count_data
```

```python
#Period 변수 value count 순서대로 정렬 : sort_index()

count2 = df['Period'].value_counts().sort_index()
count2
```

```python
#Period percent 로 표현하기

percent2 = count2/sum(count2) *100
percent2
```

```python
# 누적 count , CumCnt (Cumulative counts) 계산 : cumsum

cumcnt2 = np.cumsum(count2)
cumcnt2
```

```python
# 누적 percent-> cumsum

cumpct2 = np.cumsum(percent2)
cumpct2
```

```python
count_data2 = pd.DataFrame({'Count' : count2 , 'Percent' : percent2, 
                            'Cumcnt' : cumcnt2, 'Cumpct' : cumpct2})

#columns의 name 생성
count_data2.columns.name = 'Period'
count_data2
```

```python
###누적확률 값 계산
mu = 115
sigma = 20
x = 135
prob = stats.norm.cdf(x,mu,sigma)

print("{0}분 이상 걸리는 A/S 작업 비율: {1:.1f}%".format(x,(1-prob)*100))
```

```python
prob = stats.norm.cdf(180,175,5)
((1-prob)*100).round(1)
```

```python
prob = stats.norm.cdf(15,12,3)
prob1 = stats.norm.cdf(12,12,3)
((prob-prob1)*100).round(1)
```

```python
mu = 12
sigma = 3
x2 = 9
prob2 = stats.norm.cdf(x2, mu, sigma)
print("1) {0}km 이상 {1}이하를 달릴 확률은 {2:.1f}%에 해당함".format(x2,mu,(0.5-prob2)*100))
```

```python
t = 1.53 # t값
df = 5 #자유도
prob = stats.t.cdf(t,df)
print("P(T<=t):{:.3f}%".format(prob*100))
```

```python
chisq = 10
df = 30
prob = stats.chi2.cdf(chisq, df)
print('P(X<={0}): {1:.4f}'.format(chisq, prob))
```

```python
f = 2.0
dfnum = 20
dfden = 15

prob =stats.f.cdf(x=f, dfn = dfnum, dfd = dfden)
print("P(X<=F):{:.3f}".format(prob))
```
