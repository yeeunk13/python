```python
class Set:
    def __init__(self, val = []):
        self.val = []
        if len(val) != 0:
            for i in val:
                if i not in self.val:
                    self.val.append(i)
    def __str__(self):
        printing = str(self.val)[1:-1]
        return '{' + printing + '}'
    
    def add(self, elem):
        if elem not in self.val:
            self.val.append(elem)
            
    def discard(self, elem):
        if elem in self.val:
            self.val.remove(elem)
    
    def clear(self):
        self.val = []
        
    def __len__(self):
        return len(self.val)
    
    def __contains__(self,elem):
        if elem in self.val:
            return True
        else:
            return False
    
    def __or__(self, other):
        nums = []
        for i in other.val:
            if i not in self.val:
                nums.append(i)
        nums = self.val + nums
        return Set(nums)
    
    def __and__(self, other):
        nums = []
        for i in self.val:
            if i in other.val:
                nums.append(i)
        return Set(nums)
    
    def __sub__(self, other):
        nums = []
        for i in self.val:
            if i not in other.val:
                nums.append(i)
        return Set(nums)
    
    def __le__(self, other):
        for i in self.val:
                if i not in other.val:
                    return False
        return True
    
    def __ge__(self, other):
        for i in other.val:
            if i not in self.val:
                return False
        return True
    
    def __ior__(self,other):
        for i in other.val:
            if i not in self.val:
                self.val.append(i)
        return self
    
    def __iand__(self, other):
        for i in self.val:
            if i not in other.val:
                self.val.remove(i)
        return self
    
    def __isub__(self, other):
        for i in self.val:
            if i in other.val:
                self.val.remove(i)
        return self
```

```python
a= Set([1,2,3,4])
b= Set([1,2,3,4])
c= Set()

print(a)
print(b)
print(c)

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()


print(a | b)
print(a & b)
print(a - b)

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
